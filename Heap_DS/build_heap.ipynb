```python
from collections import Counter
import heapq
def buildheap(arr,n):
    startindex = (n // 2) - 1      # Starting swapping from last level - 1

    for i in range(startindex,-1,-1):       # Moving backwards
        heapq.heapify(arr,n,i)
        

```

```python
arr = [12,32,43,45,65,12,32,43]
n = 5
buildheap(arr,n)
```

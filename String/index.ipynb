```python
n = 'Amanetu'
count = 0
vowel = 'aeiouAEIOU'

for i in range(len(n)):
    for j in range(len(vowel)):
        if n[i] == vowel[j]:
            count += 1
print(count)
```

```python
n = "Krishnaa"
half = len(n) // 2

# Split into two halves
first_half = n[:half]
second_half = n[half:]

# Reverse second half
n = first_half + second_half[::-1]
print(n)
```

```python
n = "Krishnaa"
result = ""

for i in range(len(n)):
    if i == 2:  
        # reverse the slice from index 2 to 4 ("ish" â†’ "hsi")
        result += n[2:5][::-1]
    elif i in (3, 4):  
        # skip indices already handled by the slice
        continue
    else:
        result += n[i]

print(result)
```

```python
n = "Krishnaa"
count = 0

for i in range(len(n)):
    if i == 0:  # first char check
        if n[i] != n[i+1]:
            count += 1
    elif i == len(n) - 1:  # last char check
        if n[i] != n[i-1]:
            count += 1
    else:  # middle chars
        if n[i] != n[i-1] and n[i] != n[i+1]:
            count += 1

print(count)
```

```python
str1 = 'aman'
str2 = 'mana'

str1_sorted = sorted(str1)
str2_sorted = sorted(str2)

if str1_sorted == str2_sorted:
    print(True)
else:
    print(False)
```

```python
strr = "Aman"
freq = {}

# count frequencies
for ch in strr:
    freq[ch] = freq.get(ch, 0) + 1

# find max frequency character
max_char = max(freq, key=freq.get)  # key with highest value
print(f"Character: {max_char}, Frequency: {freq[max_char]}")
```

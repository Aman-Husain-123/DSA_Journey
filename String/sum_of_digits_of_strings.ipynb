```python
# Step 1: Character to Position Conversion
s = 'aman'
num_string = ""
for char in s:
    position = ord(char) - ord('a') + 1  # 'a' = 1, 'b' = 2, ..., 'z' = 26
    num_string += str(position)

print(num_string)
```

```python
# Step 2: Digit Sum Transformation (k times)
k = 2
for i in range(k):
    digit_sum = 0
    for digit in num_string:
        digit_sum += int(digit)
    num_string = str(digit_sum)  # Convert back to string
```

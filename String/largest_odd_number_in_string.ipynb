```python
num = "4206"
max_strr = float('-inf')
found_odd = False  # Flag to check if any odd substring exists

for i in range(len(num)):
    for j in range(i + 1, len(num) + 1):
        current_num = int(num[i:j])
        if current_num % 2 != 0:
            if current_num > max_strr:
                max_strr = current_num
                found_odd = True

# After loops, check if any odd substring was found
    if found_odd:
        print(str(max_strr))
    else:
        print('')  # print empty string if condition not satisfied
```

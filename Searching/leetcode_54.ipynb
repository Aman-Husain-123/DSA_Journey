```python
def search(nums):
    n = len(nums)
    for i in range(n):
        if (i == 0 and nums[i] != nums[i+1]) or \
           (i == n-1 and nums[i] != nums[i-1]) or \
           (0 < i < n-1 and nums[i] != nums[i-1] and nums[i] != nums[i+1]):
            return nums[i]

nums = [3,3,7,7,10,11,11]
print(nums)
print(search(nums))  # Output: 10
```

```python
def b_search(arr, target):
    lo = 0
    hi = len(arr) - 1

    while lo <= hi:
        mid = lo + (hi - lo) // 2
        if arr[mid] == target:
            return arr[mid]  # Target found
        elif target > arr[mid]:
            # The variable is 'lo', not 'low'.
            # We need to update the index, not use the array value.
            lo = mid + 1
        else:
            # The variable is 'hi', not 'high'.
            hi = mid - 1
    return -1 # Target not found

arr = [21, 32, 54, 656, 878, 989]
target = 32
print(b_search(arr, target))
```

```python
def find_first_and_last(arr, target):
    """
    Finds the first and last positions of a target in a sorted array
    using a single function.
    """
    first_pos = -1
    last_pos = -1

    # -----------------------------------------------
    ## Search 1: Find the first occurrence
    # -----------------------------------------------
    lo, hi = 0, len(arr) - 1
    while lo <= hi:
        mid = lo + (hi - lo) // 2
        if arr[mid] == target:
            first_pos = mid  # Found a match, store it
            hi = mid - 1     # BUT, keep searching LEFT for the absolute first one
        elif arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1

    # If the element was never found, first_pos will still be -1.
    # We can stop here and return.
    if first_pos == -1:
        return [-1, -1]

    # -----------------------------------------------
    ## Search 2: Find the last occurrence
    # -----------------------------------------------
    lo, hi = 0, len(arr) - 1 # Reset pointers for the new search
    while lo <= hi:
        mid = lo + (hi - lo) // 2
        if arr[mid] == target:
            last_pos = mid   # Found a match, store it
            lo = mid + 1     # BUT, keep searching RIGHT for the absolute last one
        elif arr[mid] < target:
            lo = mid + 1
        else:
            hi = mid - 1
            
    return [first_pos, last_pos]

# --- Test Cases ---
arr = [2, 2, 5, 5, 9, 15, 18, 21, 32]
x = 5
print(f"Array: {arr}")
print(f"First and last positions of {x}: {find_first_and_last(arr, x)}")
# Expected Output: First and last positions of 5: [2, 3]

print("-" * 20)

arr2 = [1, 2, 3, 3, 3, 3, 4, 5, 6]
x2 = 3
print(f"Array: {arr2}")
print(f"First and last positions of {x2}: {find_first_and_last(arr2, x2)}")
# Expected Output: First and last positions of 3: [2, 5]
```

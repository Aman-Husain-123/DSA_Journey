```python
arr = [0,1,2,3,4,5,7]
count = -1

for i in range(len(arr)):
    if i != arr[i]:
        count = i
        break

if count == -1:
    print("No missing element")
else:
    print(count)
```

```python
# Better solution using hashing

arr = [0,1,2,3,4,5,7]
hash_arr = [0] * (len(arr) + 1)

for i in range(len(arr)):
    hash_arr[arr[i]] = 1

for i in range(len(hash_arr)):
    if hash_arr[i] == 0:
        print(i)
        break
```

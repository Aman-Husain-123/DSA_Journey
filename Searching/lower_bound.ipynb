```python
# Lower Bound Using Linear Search
arr = [1,2,4,5,9,15,18,21,24]
n = 9
x = 12
for i in range(0,n):
    if arr[i] > x:
        print(arr[i - 1])
        break
```

```python
# Upper Bound Using Linear Search
arr = [1,2,4,5,9,15,18,21,24]
n = 9
x = 12
for i in range(0,n):
    if arr[i] > x:
        print(arr[i])
        break
```

```python
# Lower Bound Using Binary Search
arr = [1,2,4,5,9,15,18,21,24]
n = 9
x = 12
lo = 0
hi = n - 1
flag = False
while lo <=  hi:
    mid = lo + (hi - lo) // 2
    if arr[mid] == x:
        flag = True
        print(arr[mid - 1])
    elif arr[mid] < x:
        lo = mid + 1
    else:
        hi = mid - 1

if flag == False:
    print(arr[hi])
```

```python
arr = [1, 2, 4, 5, 9, 15, 18, 21, 24]
n = 9
x = 21
lo = 0
hi = n - 1
flag = False

# Perform binary search
while lo <= hi:
    mid = lo + (hi - lo) // 2
    if arr[mid] == x:
        # If x is found, check the next element
        if mid + 1 < n:
            print(arr[mid + 1])  # Print the next element as upper bound
        else:
            # If x is the last element, there is no upper bound
            print("No upper bound found (x is the last element)")
        flag = True
        break
    elif arr[mid] < x:
        lo = mid + 1
    else:
        hi = mid - 1

# If no exact match was found, return the first element greater than x
if not flag:
    if lo < n:
        print(arr[lo])  # Print the first element greater than x
    else:
        print("No upper bound found (x is larger than the largest element)")
```

```python
def leftSmaller(arr):
        n = len(arr)
        res = [-1] * n
        
        for i in range(n):
            for j in range(i - 1, -1, -1):  # move backwards
                if arr[j] < arr[i]:
                    res[i] = arr[j]
                    break
        
        return res
```

```python
leftSmaller([4,5,2,10,8])
```

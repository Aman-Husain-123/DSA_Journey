```python
def is_valid(s):
    """
    TC :- O(n)
    SC :- O(n)
    """

    stack = []
    mapping = {
        ')' : '(',
        ']' : '[',
        '}' : '{'
    }

    for char in s:
        if char in mapping:
            if not stack or stack[-1] != mapping[char]:
                return False
            else:
                stack.pop()
        else:
            stack.append(char)

    return len(stack) == 0
```

```python
ex1 = '{{{{{}}}}'
is_valid(ex1)
```

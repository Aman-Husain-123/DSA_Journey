```python
strr = "aman"
stack = []

lenn = len(strr)

if lenn == 0:
    print("String is empty")
else:
    for i in range(lenn):
        stack.append(strr[i])

"".join(stack[::-1])

# TODO: TC :- O(n)
# TODO: SC :- O(n)
# Steps 1 :-  Declare Empty Stack
# Steps 2 :-  Iterate to string
# Steps 3 :-  Pushing each char to stack
# Steps 4 :-  Pushing each char to stack
# Steps 5 :-  Reverse the stack
# Steps 6 :-  Again convert it into string
```

```python
# Using Push And Pop Functionality
def reverse_string(string):

    """
    TC :- O(n)
    Sc :- O(1)
    """
    stack = []
    
    # Pushing into stack linearly
    for char in string:
        stack.append(char)
    
    # Popping from stack from the end and storing it into another stack for reversing
    result = []
    while stack:
        result.append(stack.pop())
    
    return "".join(result)
```

```python
ex1 = "Aman"
ex2 = ""
reverse_string(ex2)
```

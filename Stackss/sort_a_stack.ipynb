### `Sort a stack using another stack`
- Given a stack of integers, sort it in ascending order (smallest on top) using only another stack such as auxillary storage. You can only use standard stack operations:

        - original_stack = [3,1,4,2]   # 2 is on top, 3 is at bottom
        - sorted_stack = [4,3,2,1]     # 1 is on top, 4 is at bottom

```python
stt = [20,16,8,4,5]
helper = []
print(stt)

while stt:
    val = stt.pop()
    while helper and val < helper[-1]:
        stt.append(helper.pop())
    
    helper.append(val)

while helper:
    stt.append(helper.pop())

print(stt)
print(helper)
```

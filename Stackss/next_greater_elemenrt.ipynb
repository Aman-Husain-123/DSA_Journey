```python
def next_greater_element(nums):
    result = [-1] * len(nums)
    temp = []
    for (idx,val) in enumerate(nums):
        while temp and val > temp[-1][1]:
            i,v = temp.pop()
            result[i] = val
        temp.append((idx,val))
    
    return result
```

### `Optimal Approach`

```python
# TC :- O(2N)
# SC :- O(N) + O(N)

def next_greater(nums):
    n = len(nums)
    res = [-1] * n
    stack = []
    for i in range(n - 1, -1, -1):
        while stack and stack[-1] <= nums[i]:
            stack.pop()
        res[i] = stack[-1] if stack else -1
        stack.append(nums[i])
    return res
```

```python
next_greater([1,2,32,1,244])
```

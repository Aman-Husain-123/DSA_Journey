```python

def isHappy(n):
        seen = set()
        
        while n != 1:
            if n in seen:
                return False
            seen.add(n)
            
            n = sum(int(digit) ** 2 for digit in str(n))
            print(n)
        return True
```

```python
n = 2
print(isHappy(n))  # Ou tput: True
```

```python
nums = [1,-3,2,3,-4]
max_sum = 0
n = len(nums)

for i in range(n):
    summ = 0
    for j in range(i+1,n):
        summ = abs(nums[j] + nums[i])
        max_sum = max(summ,max_sum)

max_sum
```

```python
nums = [1, -3, 2, 3, -4]
n = len(nums)

max_sum = nums[0]

for i in range(n):
    summ = 0
    for j in range(i, n):
        summ += nums[j]          # accumulate subarray
        max_sum = max(max_sum, abs(summ))

print(max_sum)
```

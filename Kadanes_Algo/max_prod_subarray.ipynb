```python
# TODO: Brute Approach TC :- O(n^2)  Sc :- O(1)
arr = [-2,1,-3,4,-1,2]
n = len(arr)
max_prod = arr[0]

for i in range(n):
    prod = 1
    for j in range(i,n):
        prod *= arr[j]
        max_prod = max(max_prod,prod)
```

```python
# TODO: Optimal Approach 
arr = [2,3,-2,4]
n = len(arr)
min_ending = arr[0]
max_ending = arr[0]
res = arr[0]

for i in range(1,n):
    v1 = arr[i]
    v2 = min_ending * arr[i]
    v3 = max_ending * arr[i]

    max_ending = max(v1,max(v2,v3))
    min_ending = min(v1,min(v2,v3))

    res = max(res,max(max_ending,min_ending))

res
```

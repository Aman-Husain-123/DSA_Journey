```python
# MAX
arr = [1, -2, 3, -2]

best_ending = arr[0]
max_ans = arr[0]

for i in range(1, len(arr)):
    v1 = best_ending + arr[i]
    v2 = arr[i]
    best_ending = max(v1, v2)
    max_ans = max(max_ans, best_ending)

print(max_ans)  # 3
```

```python
# MIN
best_ending_min = arr[0]
min_ans = arr[0]

for i in range(1, len(arr)):
    v1 = best_ending_min + arr[i]
    v2 = arr[i]
    best_ending_min = min(v1, v2)
    min_ans = min(min_ans, best_ending_min)

print(min_ans)  # -2
```

```python
total_sum = sum(arr)
print(total_sum)

# If all numbers are negative
if max_ans < 0:
    print(max_ans)
else:
    print(max(max_ans, total_sum - min_ans))
```

```python
arr = [0,0,1]
count = 0

for i in range(len(arr) - 1):
    if arr[i] == 1:
        count += 1

if count > 0:
    print("Hard")
else:
    print("Easy")
```

```python
arr = [1,2,1,0,0,1]

for i in range(0,len(arr) - 1):
    for j in range(i + 1,len(arr)):
        if arr[i] > arr[j]:
            arr[i],arr[j] = arr[j],arr[i]
            
print(arr)
```

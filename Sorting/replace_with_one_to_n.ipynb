```python
arr = [19, 12, 23, 8, 16]
n = len(arr)
v = [0] * n  # visited array

sorted_arr = []

for i in range(n):
    min_val = float('inf')
    mindx = -1
    for j in range(n):
        if v[j] == 1:  # skip visited
            continue
        if arr[j] < min_val:
            min_val = arr[j]
            mindx = j
    
    v[mindx] = 1
    sorted_arr.append(arr[mindx])

# print sorted array
for num in sorted_arr:
    print(num)
```

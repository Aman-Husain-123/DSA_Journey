```python
arr = [3,3,3,1,2,1,1,2,3,3,4]
maxLen = float('-inf')
n = len(arr)

for i in range(n):
    stt = set()
    for j in range(i,n):
        stt.add(arr[j])
        if len(stt) <= 2:
            maxLen = max(maxLen,j-i+1)
        else:
            break
```

```python
# Using HasTable And Sliding Window
fruits = [0,1,2,2]
n = 0
l = 0
maxLen = 0
mpp = {} 
k = 2

while n < len(fruits):
    
    # Add Current Fruit
    mpp[fruits[n]] = mpp.get(fruits[n],0) + 1

    # Shrink window if unique fruits > k
    while len(mpp) > k:
        mpp[fruits[l]] -= 1
        if mpp[fruits[l]] == 0:
            del mpp[fruits[l]]
        
        l += 1
    
    # Update max length
    maxLen = max(maxLen,n-l+1)
    n += 1

print(maxLen)
```

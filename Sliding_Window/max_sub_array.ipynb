```python
# Sliding Window
arr = [100,200,300,400]
k = 2
n = len(arr) # 6
max_sum = float('-inf')
left = 0

for right in range(n):
    curr_sum = 0
    while right < n and right < k:
        curr_sum += arr[right]
        right += 1
    max_sum = max(max_sum, curr_sum)

    left += 1
print(max_sum)
```

```python
arr = [100,200,300,400]
k = 2
n = len(arr) # 6
max_sum = float('-inf')
left = 0
curr_sum = 0

for right in range(n):
    curr_sum += arr[right]
    if right >= k - 1:
        max_sum = max(max_sum, curr_sum)
        curr_sum -= arr[left]
        left += 1

print(max_sum)
```

```python
arr = [100,200,300,400]
k = 2
n = len(arr) # 6
max_sum = float('-inf')
summ = 0
maxIdx = 0

# Initial Window Sum of size K(2)
while maxIdx < k and maxIdx < n:
    summ += arr[maxIdx]
    maxIdx += 1

max_sum = summ  # Initial max_sum :- 300

for i in range(1, n - k + 1):
    summ = summ - arr[i - 1] + arr[i + k - 1]
    max_sum = max(max_sum, summ)

print(max_sum)
```

```python
arr = [-8, 2, 3, -6, 10]
k = 2
result = []
index = 0
n = len(arr)

for i in range(n - k + 1):
    found = False
    for j in range(i, i + k):
        if arr[j] < 0:
            result.append(arr[j])
            found = True
            break
    if not found:
        result.append(0)
print(result)
```

```python
from collections import deque

arr = [-8, 2, 3, -6, 10]
k = 2

dq = deque()
result = []
N = len(arr)

for i in range(N):
    # Add current element index if negative
    if arr[i] < 0:
        dq.append(i)

    # Remove out-of-window elements
    while dq and dq[0] < i - k + 1:
        dq.popleft()

    # Window formed (âœ… outside while loop)
    if i >= k - 1:
        result.append(arr[dq[0]] if dq else 0)

print(result)
```

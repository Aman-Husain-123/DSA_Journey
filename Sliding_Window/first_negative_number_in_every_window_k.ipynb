```python
arr = [2, -3, 4, 4, -7, -1, 4, -2, 6]
k = 3
n = len(arr)
ans = []

for start in range(n - k + 1):
    found = False
    for j in range(start, start + k):
        if arr[j] < 0:
            ans.append(arr[j])
            found = True
            break
    if not found:
        ans.append(0)

print(ans)
```

```python
arr = [2,-3,4,4,-7,-1,4,-2,6]
n = len(arr)
k = 4
p = -1
ans = [0] * [n - k + 1]

#
for i in range(k):
    p = i
    break

if p == -1:
    ans[0] = 1
else:
    ans[0] = arr[p]

# Sliding Window
i = 1
j = k
while j < n:
    if p >= i:
        ans[i] = ans[p] 
```

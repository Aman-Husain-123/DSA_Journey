```python
# Brute Approach
arr = [2,3,1,2,4,3]
target = 7

n = len(arr)
min_length = float('inf')

for i in range(n):     # run till n times
    count = 0
    current_sum = 0
    for j in range(i,n):    # in worst case runs (n - i) times
        current_sum += arr[j]
        count += 1
        if current_sum >= target:
            min_length = min(min_length, count)
            break
if min_length == float('inf'):
    print(0)
else:
    print(min_length)
```

```python
# write the code even for the empty case
# TC :- O(n^2)
# SC :- O(256) => O(1)
s = ' abc'
max_length = 0
for i in range(len(s)):
    sett = set()
    for j in range(i, len(s)):
        if s[j] in sett:
            break
        sett.add(s[j])
    lenn = j - i + 1
    max_length = max(max_length, lenn)
```

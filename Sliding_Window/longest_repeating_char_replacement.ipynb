```python
# Optimizing it from O(N^2) to O(N)
# TC: O(N)
# SC: O(26) ~ O(1)
s = "AABABBA"
max_length = 0
n = len(s)          # Length of the string
k = 2               # Maximum replacements allowed
left = 0            # Left pointer of the sliding window
hash_map = {}       # To store the frequency of characters in the current window
max_freq = 0        # To keep track of the maximum frequency of a single character in the current window
for right in range(n):                                   # Expand the window by moving the right pointer
    hash_map[s[right]] = hash_map.get(s[right], 0) + 1   # Add the right character to the hash_map
    max_freq = max(max_freq, hash_map[s[right]])         # Update the max frequency in the current window
    window_length = right - left + 1                     # Calculate the current window length

    while window_length - max_freq > k:                  # If we have more than k replacements needed        
        hash_map[s[left]] -= 1                           # Remove the left character from the hash_map
        left += 1                                        # Shrink the window from the left    
        window_length = right - left + 1                 # Update the window length after shrinking

    max_length = max(max_length, window_length)          # Update the maximum length found
max_length
```

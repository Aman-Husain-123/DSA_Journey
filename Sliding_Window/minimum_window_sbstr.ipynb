```python
s = "ADOBECODEBANC"
t = "ABC"

m = len(s)     # length of source string
n = len(t)     # length of target string

minLen = float("inf")   # to store minimum window length
s_idx = 0               # starting index of minimum window

# Loop over all possible starting indices of s
for i in range(m):

    hashh = {}          # frequency map for characters in t
    count = 0           # count of matched characters

    # Build frequency map of t
    for ch in t:
        hashh[ch] = hashh.get(ch, 0) + 1

    # Try to extend substring from index i
    for j in range(i, m):

        # If current character is needed
        if s[j] in hashh:
            if hashh[s[j]] > 0:
                count += 1
            hashh[s[j]] -= 1

        # If all characters of t are matched
        if count == n:
            # Update minimum length
            if (j - i + 1) < minLen:
                minLen = j - i + 1
                s_idx = i
            break   # no need to extend further for this i

# Final result
if minLen == float("inf"):
    print("")
else:
    print(s[s_idx:s_idx + minLen])
```

```python
# Optimal Approach
s = "ADOBECODEBANC"
t = "ABC"

m = len(s)
n = len(t)

# Edge case
if n > m:
    print("")
    exit()

# Build frequency map for t
t_freq = {}
for ch in t:
    t_freq[ch] = t_freq.get(ch, 0) + 1

window = {}              # current window character frequencies
have = 0                 # number of chars that met required freq
need = len(t_freq)       # number of unique chars needed

left = 0
minLen = float("inf")
s_idx = 0

# Expand window
for right in range(m):

    ch = s[right]
    window[ch] = window.get(ch, 0) + 1

    # Check if current char meets requirement
    if ch in t_freq and window[ch] == t_freq[ch]:
        have += 1

    # Shrink window when valid
    while have == need:

        # Update minimum window
        if (right - left + 1) < minLen:
            minLen = right - left + 1
            s_idx = left

        # Remove left character
        window[s[left]] -= 1
        if s[left] in t_freq and window[s[left]] < t_freq[s[left]]:
            have -= 1

        left += 1

# Result
if minLen == float("inf"):
    print("")
else:
    print(s[s_idx:s_idx + minLen])
```

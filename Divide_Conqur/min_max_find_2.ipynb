```python
def min_max(arr,i,j):
    # Small problem
    # Single element
    if i == j:
        min_val = arr[i]
        max_val = arr[i]
    
    # Two element condition
    elif i == j - 1:
        if arr[i] < arr[j]:
            max_val = arr[j]
            min_val = arr[i]
        else:
            max_val = arr[i]
            min_val = arr[j]
    
    # Big Problem
    else:
        # Divide And Conque Approach
        # 1. Divide
        mid = i + (j - i) // 2

        # Recursion
        # 2. Recursion -> Conquer
        min_1,max_1 = min_max(arr,i,mid)
        min_2,max_2 = min_max(arr,mid + 1,j)

        # Combine
        min_val = min_1 if min_1 < min_2 else min_2 
        max_val = max_1 if max_1 > max_2 else max_2 
    
    return(min_val,max_val)
```

```python
arr = [12,32,21,1,22,34]
i = 0
j = len(arr) - 1
min_element,max_element = min_max(arr,i,j)
print(min_element,max_element)
```

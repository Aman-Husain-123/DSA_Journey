```python
def powerofelement(a,n):
    # Small problem
    if n == 1:
        return a
    elif n == 0:
        return 1
    else:
        # Big Problem(Divide And Conquer Approach)
        # 1. Divide
        mid = n // 2

        # 2. Conquer
        b = powerofelement(a,mid)

        # 3. Combine
        result = b * b

        # Condition checking for even and odd
        if n % 2 == 0:
            return result
        else:
            return result * a
```

```python
# Driver Code
a = 2
n = 17

# Function Calling
result = powerofelement(a,n)
print(f"Power of element is:",result)
```

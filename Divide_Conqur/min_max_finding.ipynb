```python
def findmaxandmin(arr, i, j):
    if i == j:
        min_val = arr[i]
        max_val = arr[i]
    elif i == j - 1:
        if arr[i] < arr[j]:
            max_val = arr[j]
            min_val = arr[i]
        else:
            max_val = arr[i]
            min_val = arr[j]
    else:
        mid = i + (j - i) // 2
        min1, max1 = findmaxandmin(arr, i, mid)
        min2, max2 = findmaxandmin(arr, mid + 1, j)

        min_val = min1 if min1 < min2 else min2
        max_val = max1 if max1 > max2 else max2

    return (min_val, max_val)
```

```python
arr = [4, 10, 3, 5, 1, 8]
print("Minimum and Maximum:", findmaxandmin(arr, 0, len(arr) - 1))
```

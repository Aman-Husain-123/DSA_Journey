```python
# TODO: Using Recursion

def removeChar(ans, original,idx):
    # base case
    if len(original) == idx:
        print(ans)
        return
    
    ch = original[idx]
    if ch == 'h':
        removeChar(ans, original,idx + 1)      # skip 'h'
    else:
        removeChar(ans + ch, original,idx + 1) # keep character


# Driver code
s = "physics wallah"
removeChar("",s,0)
```

```python
# TODO: Using Recursion

def removeChar(ans, original,idx):
    # base case
    if len(original) == idx:
        print(ans)
        return
    
    ch = original[idx]
    if ch == 1:
        removeChar(ans, original,idx + 1)      # skip 'h'
    else:
        removeChar(ans + [ch], original,idx + 1) # keep character


# Driver code
s = [1,1,2,3,2,1,2]
removeChar([],s,0)
```

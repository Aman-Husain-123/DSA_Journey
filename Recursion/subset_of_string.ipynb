```python
def printsubset(ans,original):
    if original == "":
        print(ans,end = " ")
        return
    ch = original[0]
    printsubset(ans+ch,original[1:])
    printsubset(ans,original[1:])
```

```python
str = "abc"
printsubset("",str)
```

```python
def printsubset(arr, n, idx, ans):
    if idx == n:
        for i in range(len(ans)):
            print(ans[i], end=" ")
        print()
        return

    # exclude current element
    printsubset(arr, n, idx + 1, ans)

    # include current element
    ans.append(arr[idx])
    printsubset(arr, n, idx + 1, ans)

    # ðŸ”¥ backtrack â€” remove the last element
    ans.pop()
```

```python
str = [1,2,3]
n = len(str)
v = []
printsubset(str,n,0,v)
```

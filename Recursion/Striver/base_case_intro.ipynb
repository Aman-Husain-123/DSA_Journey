#### `Base Case`

```python
def printt(n,idx):
    # TODO: Base Case
    if idx == n:
        return
    print("Aman")
    idx += 1
    printt(n,idx)
```

```python
printt(2,0)
```

```python
def printt(count = 0):
    # TODO: Base Case
    if count == 4:
        return
    print(count)
    printt(count + 1)
```

```python
def func(n,idx = 1):
    ans = 0
    if idx > n:
        return 0
    ans += idx
    ans += func(n, idx + 1)    # accumulate recursive result
    return ans
```

```python
print(func(5))
```

```python
def d(i,sum):
    if i < 1:
        print(sum)
        return
    
    d(i - 1,sum + i)
```

```python
d(5,0)
```

```python
def func(n):
    if n == 0:
        return 0
    return n + func(n - 1)

func(6)
```

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n - 1)
```

```python
fact(4)
```

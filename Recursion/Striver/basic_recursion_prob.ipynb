```python
def printt(idx,n):              ### TC And SC :- O(n)
    if idx > n:
        return
    print("Aman")
    printt(idx + 1,n)
```

```python
n = int(input("Enter Number:"))
printt(1,n)
```

```python
def printt(n):
    if n < 1:
        return
    
    print(n)
    printt(n - 1)
```

```python
printt(4)
```

```python
# TODO: Print from 1 - N without using i + 1 by backtracking
```

```python
def func(i,n):
    if i < 1:
        return
    func(i - 1,n)
    print(i)
```

```python
func(3,3)
```

```python
def func(n,i):
    if i < 1:
        return
    print(i)
    func(n,i - 1)
```

```python
strr = "DaD"
i = 0
j = len(strr) - 1

flag = True
while i < j:
    if strr[i] != strr[j]: 
            flag = False
            break # Exit immediately on mismatch
            
        # Move pointers inward for the next comparison
    i += 1
    j -= 1


if flag == False:
    print("Not a palindrome")
else:
    print("Palindrome")
```

```python
# TODO: Using Recursion
def f(strr,i = 0):
    mid = len(strr) // 2
    n = len(strr) - 1
    if i >= mid:
        return True
    
    if strr[i] != strr[n - i]:
        return False
    return f(strr,i + 1)
```

```python
result = f("DAD",0)
if result == True:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
def printS(ind,ds,s,sum,arr,n):
    if ind == n:
        if s == sum:
            for i in range(len(ds)):
                print(ds[i],end = " ")
            print()
        return 
    
    ds.append(arr[ind])
    s += arr[ind]
    printS(ind+1,ds,s,sum,arr,n)
    s -= arr[ind]
    ds.pop()
    printS(ind + 1,ds,s,sum,arr,n)
```

```python
arr = [1,2,1]
n = 3
sum = 2
arr1 = []
printS(0,arr1,0,sum,arr,n)
```

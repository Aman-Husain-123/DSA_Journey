```python
arr = [1,2,3,4,2]
start = 0
end = len(arr) - 1


while start < end:
    arr[start],arr[end] = arr[end],arr[start]
    start += 1
    end -= 1
print(arr)
```

```python
# TODO: Using recursion two pointer approach
def reverse(a,l,r):
    if l >= r:
        return
    a[l],a[r] = a[r],a[l]
    reverse(a,l + 1,r - 1)
    
```

```python
arr = [1,2,3,4,2]
n = len(arr) - 1
print("Original array:", arr)
reverse(arr,0,n)
print("Reversed array:", arr)
```

```python
# TODO: Using recursion one pointer approach
from typing import List
def reverse(a,i) -> List[int]:
    n = len(a)
    mid = len(a) // 2
    if i >= mid:
        return
    a[i],a[n - i - 1] = a[n - i - 1],a[i]
    reverse(a,i + 1)
    
```

```python
arr = [100,200,300]
n = len(arr) - 1
print("Original array:", arr)
reverse(arr,0)
print("Reversed array:", arr)
```

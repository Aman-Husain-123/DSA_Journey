```python
def print_subsequence(index,current_subsequence,array,n):
    # Base Condition: When the index reaches the size of the array (n),
    # it means all elements have been processed. The current_subsequence is complete.
    if index == n:
        # Print the current subsequence
        if not current_subsequence:
             # Prints "{}" for the empty subsequence
            print('[]')
        else:
            # Prints space-separated element
            print(*current_subsequence)
        return
    
    # --- 1. TAKE / PICK Condition ---
    # Take or pick the element at the current index and add it to the subsequence.
    current_subsequence.append(array[index])

    # Recurse: Move to the next index (index + 1) with the updated subsequence.
    print_subsequence(index + 1,current_subsequence,array,n)

    
    # Backtrack: Remove the element added in this step. This is crucial
    # to maintain the correct state for the "NOT TAKE" condition below.
    current_subsequence.pop()

    
    # --- 2. NOT TAKE / NOT PICK Condition ---
    # Do not take the element at the current index.
    # We call the function for the next index (index + 1) without adding the current element.
    # The 'current_subsequence' list remains unchanged from before the backtracking step.
    print_subsequence(index + 1,current_subsequence,array,n)

```

```python
# Main execution
if __name__ == "__main__":
    # The input array as defined in the C++ image
    arr = [3, 1, 2]
    n = len(arr)

    # Initialize an empty list to store the current subsequence (equivalent to C++ vector 'ds')
    data_structure = []

    # Start the recursion from the first element (index 0)
    print("Generating subsequences for:", arr)
    print_subsequence(0, data_structure, arr, n)
```

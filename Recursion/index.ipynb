```python
# Using an extra parameter
def printt(i,n) -> None:
    if i > n:
        return
    
    print(i)
    printt(i + 1,n)
```

```python
n = int(input("Enter Input:"))
printt(1,n)
```

```python
def sum_to_n_corrected(current_sum, n):
    if n == 0:
        return current_sum #  Return the final sum instead of printing
    
    #  Return the result of the next recursive call
    return sum_to_n_corrected(current_sum + n, n - 1) 

n = int(input("Enter nth term:"))
#  The caller is now responsible for printing the returned value
result = sum_to_n_corrected(0, n)
print(result)
```

```python
# Print sum from 1 to n (Return Type)
def summation(n):
    if n == 0:
        return 0
    else:
        return n + summation(n - 1)
```

```python
summation(4)
```

```python
# Calculates 'a' raised to the power 'b' using recursion.
# This implementation uses only multiplication (*) and division (/).
# No built-in power operator (**) or function (math.pow) is used.

def power(a, b):
    """
    Calculates a raised to the power of b (a^b) recursively.
    
    Args:
        a (float): The base number.
        b (int): The exponent.
        
    Returns:
        float: The result of a^b.
    """
    
    # --- Base Case 1: Exponent is 0 ---
    # a^0 = 1
    if b == 0:
        return 1.0
    
    # --- Recursive Case 1: Positive Exponent ---
    # a^b = a * a^(b-1)
    if b > 0:
        # We achieve a^b through repeated multiplication.
        return a * power(a, b - 1)
        
    # --- Recursive Case 2: Negative Exponent ---
    # a^(-b) = 1 / a^b
    if b < 0:
        # We calculate the positive power, then return its reciprocal (division).
        positive_power = power(a, -b)
        return 1.0 / positive_power


# --- Example Usage (Hardcoded) ---
base = 2.5
exponent = 3

result = power(base, exponent)
print(f"Example 1: {base} raised to the power {exponent} is {result}")

base_neg = 5
exponent_neg = -2

result_neg = power(base_neg, exponent_neg)
print(f"Example 2: {base_neg} raised to the power {exponent_neg} is {result_neg}")
```

```python
def pow(x,n) -> int:
    if n == 0:
        return 1
    else:
        return x * pow(x,n-1)
```

```python
pow(3,4)
```

```python
def pow(x,n) -> int:
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        return x * pow(x,n-1)
```

```python
pow(3,0)
```

```python
def pow(x,n) -> int:
    if n == 1:
        return x
    ans = pow(x,n / 2)
    return ans * ans
```

```python
pow(3,8)
```

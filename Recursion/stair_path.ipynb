```python
def stair(n):
    if n == 2:
        return 2
    elif n == 1:
        return 1
    else:
        return stair(n-1) + stair(n-2)
```

```python
stair(5)
```

```python
# Tree step as well
def stair(n):
    if n == 2:
        return 2
    elif n == 1:
        return 1
    elif n == 3:
        return 3
    else:
        return stair(n-1) + stair(n-2) + stair(n - 3)
```

```python
stair(5)
```

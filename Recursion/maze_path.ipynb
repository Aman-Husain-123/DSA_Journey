```python
def maze(sr,sc,er,ec):
    if sr > er or sc > ec:
        return 0
    
    if sr == er and sc == ec:
        return 1
    rightways = maze(sr,sc + 1,er,ec)
    downways = maze(sr + 1,sc,er,ec)

    totalways = rightways + downways
    return totalways


# With 2 Parameter
def maze2(row,col):
    if row < 1 or col < 1:
        return 0
    
    if row == 1 and col == 1:
        return 1
    rightways = maze2(row ,col-1)
    downways = maze2(row - 1,col)

    totalways = rightways + downways
    return totalways

def printpath(sr,sc,er,ec,s):
    if sr > er or sc > ec:
        return
    
    if sr == er and sc == ec:     # Destination Reached
        print(s)
        return
    printpath(sr,sc + 1,er,ec,s + "R")   #right
    printpath(sr + 1,sc,er,ec,s + 'D')   #down

```

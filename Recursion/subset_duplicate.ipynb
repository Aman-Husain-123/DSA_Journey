```python
def storesubset(ans, original, v, flag):
    # Base case
    if original == "":
        v.append(ans)
        return

    ch = original[0]

    # Include current character
    storesubset(ans + ch, original[1:], v, True)
    
    # Skip all consecutive same characters if flag is True
    if flag:
        idx = 1
        while idx < len(original) and original[idx] == ch:
            idx += 1
        storesubset(ans, original[idx:], v, True)


# Driver code
strr = "aab"
v = []
storesubset("", strr, v, True)

print("All stored subsets:")
for subset in v:
    print(subset)
```

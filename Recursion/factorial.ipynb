```python
# TODO:Without Recursion
def func(n):
    fact = 1
    while n > 0:
        fact *= n
        n -= 1
    return fact
```

```python
print(func(4))
```

```python
# TODO: Using Recursion
def func(n):
    # Base Case
    if n == 1:
        return 1
    elif n == 0:
        return "Not Valid"
    else:
        return n * func(n - 1)
```

```python
print(func(0))
```

```python
# Using Extra Parameter
def printt(i, n):
    if i > n:
        return
    print(i)
    printt(i + 1, n)

n = int(input("Enter Number: "))
printt(1, n)
```

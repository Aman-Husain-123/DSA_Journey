```python
def backspaceCompare(s, t) -> bool:

        def process(string):
            stack = []
            for ch in string:
                if ch == '#':
                    if stack:
                        stack.pop()
                else:
                    stack.append(ch)
            return "".join(stack)

        return process(s) == process(t)
```

```python
s = "ab#c"
t = "ad#c"

print(backspaceCompare(s,t))
```

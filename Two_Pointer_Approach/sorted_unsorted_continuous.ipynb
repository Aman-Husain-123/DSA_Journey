```python
arr = [1]
nums = sorted(arr)

for i in range(len(arr)):
    if arr[i] != nums[i]:
        small = i
        break

for j in range(len(arr)-1, -1, -1):
    if arr[j] != nums[j]:
        large = j
        break

print(large - small + 1)
```

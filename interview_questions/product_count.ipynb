```python
products = ['yellowShirt', 'redHat', 'blackShirt', 'bluePants', 
            'redHat', 'pinkHat', 'blackShirt', 'yellowShirt', 'greenPants']
```

```python
# Step : 1 Count frequencies manually
products_count = {}
for item in products:
    if item in products_count:
        products_count[item] += 1
    else:
        products_count[item] = 1
print("Frequencey dictionary:",products_count) 
```

```python
# Step 2: Find maximum frequency
max_count = 0
for product in products_count:
    if products_count[product] > max_count:
        max_count = products_count[product]

print('Max Frequency:',max_count)
```

```python
# Step 3: Collect all products with max frequency
rgd_product = []
for key,value in products_count.items():
    if value == max_count:
        rgd_product.append(key)


print("Candidates with max freq:", rgd_product)
```

```python
# Step 4: Pick the last alphabetically
result = rgd_product[0]
for product in rgd_product:
    if product > result:   # lexical comparison
        result = product

print("Final result:", result)
```

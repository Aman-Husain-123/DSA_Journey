```python
def target_sum(arr,x):

    closest_sum = abs(x * 100000)
    arr.sort()

    for i in range(len(arr) - 2):

        left = i + 1
        right = len(arr) - 1

        while left <= right:
            sumVal = arr[i] + arr[left] + arr[right]
            if(abs(x-sumVal) < abs(x - closest_sum)):
                closest_sum = sumVal
            
            elif sumVal > x:
                right -= 1
            else:
                left += 1
    
    return closest_sum
```

```python
arr = [1,2,-1,3,4]
target = 2
print(target_sum(arr,target))
```

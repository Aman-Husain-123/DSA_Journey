```python
from typing import List
def numWaterBottles(numBottles: int, numExchange: int) -> int:
        drank = numBottles
        while numBottles >= numExchange:

            # log(numBottles) to the base of exchanged

            exchanged = numBottles // numExchange
            extraEmpty = numBottles % numExchange

            drank += exchanged
            numBottles = exchanged + extraEmpty
        
        return drank
```

```python
numBottles = 9
numExchange = 3
numWaterBottles(numBottles,numExchange)
```

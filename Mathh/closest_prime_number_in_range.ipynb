```python
def find_numbers(left,right):
    primes = []
    for num in range(left, right + 1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes
```

```python
# Example usage
left = 10
right = 19
result = find_numbers(left,right)
print(result)  # Output: [11, 13, 17, 19]
```

```python
final_res = []
for i in range(len(result)):
    if (result[i] >= left and result[i] < right) and (result[i + 1] > result[i] and result[i - 1] < right):
        final_res[i].append(result[i])
```

```python
def find_two_primes(start, end):
    primes = []
    for num in range(start, end + 1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    
    # Find two primes where:
    # 1st prime >= start and < end
    # 2nd prime > 1st prime and < end
    for i in range(len(primes) - 1):
        if primes[i] >= start and primes[i] < end and primes[i+1] > primes[i] and primes[i+1] < end:
            return [primes[i], primes[i+1]]
    
    return None  # Return None if no such pair exists

# Example usage
result = find_two_primes(10, 19)
print(result)  # Output: [11, 13]
```

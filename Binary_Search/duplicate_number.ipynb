```python
nums = [1,3,4,2,2]
count_dict = {}

# Count frequencies
for num in nums:
    count_dict[num] = count_dict.get(num, 0) + 1

# print(count_dict)
# Print elements with count > 2
for item in count_dict.items():
    if item[1] > 1:
        print(item[0])
```

```python
def findFixedPoint(nums):
    """
    Binary search works ONLY if input is already sorted
    Like [0, 1, 2, 3] or [-2, 0, 1, 3, 5]
    """
    left = 0
    right = len(nums) - 1
    result = -1
    
    while left <= right:
        mid = (left + right) // 2
        
        if mid == nums[mid]:
            result = mid
            right = mid - 1  # Check left side for more matches
        elif mid < nums[mid]:
            right = mid - 1  # Search left
        else:
            left = mid + 1   # Search right
    
    return result

# Works for sorted arrays
print(findFixedPoint([0, 1, 2, 3]))      # 0 (or 1, 2, 3)
print(findFixedPoint([-2, 0, 1, 3, 5]))  # 3

# Does NOT work for unsorted
print(findFixedPoint([3, 1, 3, 4, 2]))   # WRONG!
```

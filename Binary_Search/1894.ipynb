```python
from typing import List 
def chalkReplacer(chalk: List[int], k: int) -> int:
        # TODO: 
        n = len(chalk)

        # Step1: Build prefix sum array
        prefix = [0] * n
        prefix[0] = chalk[0]
        for i in range(1,n):
            prefix[i] = prefix[i-1] + chalk[i]
        
        # Step2: Reduce k using modulo
        total = prefix[-1]
        k = k % total

        # Step3: Binary search for the student
        left, right = 0, n - 1
        result = 0

        while left <= right:
            mid = (left + right) // 2
            if prefix[mid] <= k:
                left = mid + 1
            else:
                result = mid
                right = mid - 1
        
        return result
```

```python
chalk = [5,10,15,20]
k = 14
chalkReplacer(chalk,k)
```

```python
# TODO: TC :- O(N)
def linear_Search(arr,n,target):
    first_pos,last_pos = -1,-1  
    for i in range(n):
        if arr[i] == target:
            if first_pos == -1:
                first_pos = i
            else:
                last_pos = i
    
    return first_pos,last_pos
```

```python
arr = [1,2,3,3]
target = 3
n = len(arr)
print(linear_Search(arr,n,target))
```

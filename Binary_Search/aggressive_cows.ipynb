```python
def aggressive_cows(arr, cows):
    arr.sort()    # Make sure stall positions are sorted
    left = 1      # Minimal possible distance
    right = arr[-1] - arr[0]  # Max possible distance across stalls
    best = 0
    while left <= right:
        mid = (left + right) // 2
        if can_we_place(arr, mid, cows):
            best = mid         # Current distance is possible, try for more
            left = mid + 1
        else:
            right = mid - 1    # Try for less (can't fit all cows)
    return best
```

```python
# Checks if we can place 'cows' in stalls 'arr' so that no two cows are closer than 'dist'
def can_we_place(arr, dist, cows):
    count_cows = 1        # Always place first cow at the first stall
    last = arr[0]
    for i in range(1, len(arr)):
        if arr[i] - last >= dist:
            count_cows += 1
            last = arr[i]
    return count_cows >= cows
```

```python
# Example stall arrangement and number of cows
arr = [0, 3, 4, 7, 9, 10]
cows = 4
print(aggressive_cows(arr, cows))  # Output: Maximum possible minimum distance
```

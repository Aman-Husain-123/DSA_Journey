```python
class ValueNotFoundError(Exception):
    pass

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None 

class LinkedList:
    def __init__(self, head=None):
        self.head = head
        self.size = 0
        if head is not None:
            self._initialize_size()

    def _initialize_size(self):
        count = 0
        curr = self.head
        while curr:
            count += 1
            curr = curr.next
        self.size = count

    def is_empty(self):    
        return self.head is None

    def append(self, value):

        new_node = Node(data=value)
        
        if self.is_empty():
            self.head = new_node
            self.size = 1
            print("New size of list:", self.size)
            return
        
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node
        
        self.size += 1  # Increment size properly
        print("New size of list:", self.size)

    def traverse(self):
        if self.is_empty():
            return None
        
        elements = []
        curr = self.head
        while curr:
            elements.append(str(curr.data))
            curr = curr.next
        return " -> ".join(elements)

    def check_membership(self, value):
        if self.is_empty():
            raise ValueNotFoundError(f"Value {value} not found in the list")
        
        curr = self.head
        while curr:
            if curr.data == value:
                return curr.data
            curr = curr.next

        raise ValueNotFoundError(f"Value {value} not found in the list")

    def get_size(self):
        return self.size

    def delete_at_beginning(self):
        if self.is_empty():
            raise Exception("LinkedList Is Empty")
        
        self.head = self.head.next
        self.size -= 1  # Correct decrement by 1
        print("New size of list:", self.size)

    def delete_at_end(self):
        if self.is_empty():
            raise Exception("LinkedList Is Empty")
        
        if self.head.next is None:
            self.head = None
        else:
            curr = self.head
            while curr.next.next:
                curr = curr.next
            curr.next = None
        
        self.size -= 1  # Correct decrement by 1
        print("New size of list:", self.size)
    
    def delete_at_position(self,pos):
        """
        This will delete the element at any specific position
        TC :- 
        """

        # TODO: Edge Case:1 When LL is emtpy
        if self.is_empty():
            raise Exception("LinkedList Is Empty")
        
        # TODO: Edge Case : 2  If position is invalid (pos is negative or more than the size of the list)
        if pos < 0 or pos > self.size:
            raise Exception("Invalid Position")
        
        # TODO: Edge Case : 3 Deleting first element if pos = 0
        if pos == 0:
            self.delete_at_beginning()
        
        
        curr = self.head
        for _ in range(pos - 1):
            curr = curr.next
        curr.next = curr.next.next
        self.size -= 1
        print("New size of list:", self.size)
```

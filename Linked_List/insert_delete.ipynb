```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
# Printing
def print_linked_list(head):
    current = head
    while current is not None:
        print(current.data,end = " ")
        current = current.next
```

```python
# Creating a LL
head = Node(20)
head.next = Node(30)
head.next.next = Node(40)
head.next.next.next = Node(50)

print_linked_list(head)
```

```python
def indert_at_start(head,value):
    new_node = Node(value)

    new_node.next = head
    return new_node

head = indert_at_start(head,300)
print_linked_list(head)
```

### `Add at specific position`

```python
def insert_at_pos(head,value,position):
    new_node = Node(value)

    # When element has to be add on first position
    if position == 1:
        new_node.next = head
        return new_node
    

    curr = head

    for i in range(1,position-1):
        if curr is None:
            print("Invalid Position")
            return head
        
        curr = curr.next
    
    if curr is None:
        print("Invalid Position")
        return head

    new_node.next = curr.next
    curr.next = new_node

    return head
```

### `Deleting at the start`

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
def delete_at_start(head):
    if head is None:
        print("No element in LL")
        return None

    new_head = head.next
    head.next = None
    return new_head
```

### `Delete at end` 

```python
def dele_at_end(head):
    if head is None:
        print("No element")
        return None
    
    if head.next is None:
        return None
    
    curr = head
    while curr.next.next is not None:
        curr = curr.next
    
    curr.next = None

    return head
```

```python
head = dele_at_end(head)
print_linked_list(head)
```

### `Que : 1` Deleting Every Kth Node


```python
class Node:
    def __init__(self,new_data):
        self.data = new_data
        self.next = None
```

```python
# Function to remove every kth node in the linked list
def delete_k(head,k):

    # If the list is empty or k is 0,return the head
    if head is None or k <= 0:
        return head

    curr = head
    prev = None

    count = 0

    while curr is not None:
        count += 1
 # Bypass the current node
        if prev is not None:
                prev.next = curr.next
        else:
              
                # If removing the head node
                head = curr.next
    else:
          
            # Update previous node pointer only if
            # we do not remove the node
            prev = curr

            curr = curr.next

    return head


def print_list(head):
    curr = head
    while curr is not None:
        print(curr.data, end=" ")
        curr = curr.next
    print()
```

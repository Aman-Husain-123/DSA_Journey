### `Solution`


```python
# 
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Solution:
    def addTwoNumbers(self, num1, num2):
        dummyHead = Node(-1)
        curr = dummyHead
        temp1 = num1
        temp2 = num2
        carry = 0
        
        while temp1 is not None or temp2 is not None:
            _sum = carry
            if temp1:
                _sum += temp1.data
            if temp2:
                _sum += temp2.data
            newNode = Node(_sum % 10)
            carry = _sum // 10
            
            curr.next = newNode
            curr = curr.next
            
            if temp1:
                temp1 = temp1.next
            if temp2:
                temp2 = temp2.next
        
        if carry:
            curr.next = Node(carry)
            
        return dummyHead.next

# ====== Optional: Helper code to test ======

def print_linked_list(head):
    vals = []
    while head:
        vals.append(str(head.data))
        head = head.next
    print(" -> ".join(vals))

def create_linked_list(lst):
    if not lst:
        return None
    head = Node(lst[0])
    curr = head
    for v in lst[1:]:
        curr.next = Node(v)
        curr = curr.next
    return head

if __name__ == "__main__":
    num1 = create_linked_list([2, 4, 3])  # represents 342
    num2 = create_linked_list([5, 6, 4])  # represents 465
    sol = Solution()
    result = sol.addTwoNumbers(num1, num2)  # should output 7 -> 0 -> 8 (807)
    print_linked_list(result)
```

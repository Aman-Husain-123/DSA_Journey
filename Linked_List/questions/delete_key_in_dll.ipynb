```python
class Node:
    def __init__(self,val):
        self.val = val
        self.prev = None
        self.next = None
```

```python
def deleteAllKeyDLL(head,key):
    temp = head
    while temp is not None:
        if temp.val == key:
            # If the node to be deleted is the head
            if temp == head:
                head = head.next
                if head:
                    head.prev = Node
            else:
                nextNode = temp.next
                prevNode = temp.prev

                # Fix the prev pointer of the next node, if it exists
                if nextNode:
                    nextNode.prev = prevNode
                
                # Fix the next pointer of the previous node, if it exists
                if prevNode:
                    prevNode.next = nextNode
            
            # Store the next node before deleting current node
            next_temp = temp.next
            
            # Optionally, explicitly delete the node in Python: del temp
            temp = next_temp
        else:
            temp = temp.next
    return head
```

```python
# Helper function to print DLL
def printDLL(head):
    curr = head
    while curr:
        print(curr.val, end=' <-> ' if curr.next else '')
        curr = curr.next
    print()
```

```python
# Example usage:
# DLL: 1 <-> 3 <-> 3 <-> 4 <-> 3 <-> 5; key = 3
head = Node(1)
n2 = Node(3)
n3 = Node(3)
n4 = Node(4)
n5 = Node(3)
n6 = Node(5)
head.next = n2; n2.prev = head
n2.next = n3; n3.prev = n2
n3.next = n4; n4.prev = n3
n4.next = n5; n5.prev = n4
n5.next = n6; n6.prev = n5

print("Before deletion:")
printDLL(head)
head = deleteAllKeyDLL(head, 3)
print("After deletion of 3:")
printDLL(head)
```

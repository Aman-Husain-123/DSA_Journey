```python
from typing import Optional

# Definition for singly linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution:
    def oddEvenList(self,head):
        """
        Reverse-engineer approach using array:
        1. Store odd-indexed nodes in array
        2. Store even-indexed nodes in array
        3. Replace values sequentially
        
        Time Complexity: O(n)
        Space Complexity: O(n) - array storage
        """

        # Edge Case: Empty List or single node
        if not head or not head.next:
            return head
        
        count = 0
        temp = head
        n = 2
        for i in range(ListNode):
            count += 1
            temp = temp.next
        
        temp = count - n
        temp.next = temp.next.next

```

```python
# Helper Functions
def create_linked_list(arr):
    """Create linked list from array"""
    if not arr:
        return None
    head = ListNode(arr[0])
    curr = head
    for val in arr[1:]:
        curr.next = ListNode(val)
        curr = curr.next
    return head

def print_linked_list(head):
    """Print linked list"""
    values = []
    curr = head
    while curr:
        values.append(str(curr.val))
        curr = curr.next
    return " -> ".join(values)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # Edge case: empty or single node
        if not head or not head.next:
            return head
        
        # Step 1: Count total nodes
        count = 0
        temp = head
        while temp:
            count += 1
            temp = temp.next
        
        # Step 2: Calculate position from start
        position_from_start = count - n
        
        # Step 3: Handle removing head
        if position_from_start == 0:
            return head.next
        
        if count == 0:
            return []
        
        # Step 4: Traverse to node before target
        temp = head
        for i in range(position_from_start - 1):
            temp = temp.next
        
        # Step 5: Remove the node
        temp.next = temp.next.next
        
        return head


def create_ll(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    curr = head
    for v in arr[1:]:
        curr.next = ListNode(v)
        curr = curr.next
    return head

def print_ll(head):
    vals = []
    while head:
        vals.append(str(head.val))
        head = head.next
    return " -> ".join(vals)

# Test Cases
sol = Solution()

print("Test 1: [1,2,3,4,5], n=2")
head = create_ll([1, 2, 3, 4, 5])
print(f"Before: {print_ll(head)}")
head = sol.removeNthFromEnd(head, 2)
print(f"After:  {print_ll(head)}")
print(f"Expected: 1 -> 2 -> 3 -> 5\n")

print("Test 2: [1], n=1")
head = create_ll([1])
print(f"Before: {print_ll(head)}")
head = sol.removeNthFromEnd(head, 1)
print(f"After:  {print_ll(head) if head else 'None'}")
print(f"Expected: None\n")

print("Test 3: [1,2], n=2")
head = create_ll([1, 2])
print(f"Before: {print_ll(head)}")
head = sol.removeNthFromEnd(head, 2)
print(f"After:  {print_ll(head)}")
print(f"Expected: 2\n")

print("Test 4: [1,2,3], n=3")
head = create_ll([1, 2, 3])
print(f"Before: {print_ll(head)}")
head = sol.removeNthFromEnd(head, 3)
print(f"After:  {print_ll(head)}")
print(f"Expected: 2 -> 3")
```

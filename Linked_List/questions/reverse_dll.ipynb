```python
class DNode:
    """Node for doubly linked list"""
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def append(self, data):
        """Add element at end"""
        new_node = DNode(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
    
    def print_forward(self, list_name="List"):
        """Print list from head to tail"""
        values = []
        curr = self.head
        while curr:
            values.append(curr.data)
            curr = curr.next
        print(f"{list_name} (Forward): {' <-> '.join(map(str, values))}")
    
    def print_backward(self, list_name="List"):
        """Print list from tail to head"""
        values = []
        curr = self.tail
        while curr:
            values.append(curr.data)
            curr = curr.prev
        print(f"{list_name} (Backward): {' <-> '.join(map(str, values))}")


# ===== APPROACH 1: STACK-BASED REVERSAL =====

def reverse_dll_stack(dll):
    """
    Reverse doubly linked list using stack
    
    Steps:
    1. Store all node values in stack (push from head to tail)
    2. Pop from stack and place values back into list (head to tail)
    
    Time Complexity: O(n)
    Space Complexity: O(n) - for stack
    """
    print("\n" + "=" * 70)
    print("REVERSING DOUBLY LINKED LIST - STACK APPROACH")
    print("=" * 70)
    
    if not dll.head:
        print("List is empty!")
        return dll
    
    # Step 1: Push all node data into stack
    print("\nStep 1: Pushing all elements into stack...")
    stack = []
    curr = dll.head
    position = 0
    
    while curr:
        stack.append(curr.data)
        print(f"  Push {curr.data} into stack (position {position})")
        position += 1
        curr = curr.next
    
    print(f"\nStack after pushing all elements: {stack}")
    
    # Step 2: Pop from stack and replace values in list
    print("\n\nStep 2: Popping elements from stack and replacing in list...")
    curr = dll.head
    position = 0
    
    while curr and stack:
        popped_value = stack.pop()
        print(f"  Pop {popped_value} from stack -> Replace position {position} (was {curr.data})")
        curr.data = popped_value
        position += 1
        curr = curr.next
    
    print(f"\nList after reversal (stack-based):")
    dll.print_forward("  ")
    
    return dll


# ===== APPROACH 2: OPTIMAL - SWAP POINTERS (Better Approach) =====

def reverse_dll_pointers(dll):
    """
    Reverse doubly linked list by swapping next and prev pointers
    
    Steps:
    1. Traverse the list
    2. For each node, swap its next and prev pointers
    3. Update head and tail
    
    Time Complexity: O(n)
    Space Complexity: O(1) - no extra space
    """
    print("\n" + "=" * 70)
    print("REVERSING DOUBLY LINKED LIST - POINTER SWAP (OPTIMAL)")
    print("=" * 70)
    
    if not dll.head:
        print("List is empty!")
        return dll
    
    curr = dll.head
    position = 0
    
    print(f"\nSwapping next and prev pointers for each node...\n")
    
    # Swap pointers for all nodes
    while curr:
        print(f"  Position {position}: Node({curr.data})")
        print(f"    Before: next={curr.next.data if curr.next else 'None'}, prev={curr.prev.data if curr.prev else 'None'}")
        
        # Swap next and prev
        curr.next, curr.prev = curr.prev, curr.next
        
        print(f"    After:  next={curr.next.data if curr.next else 'None'}, prev={curr.prev.data if curr.prev else 'None'}")
        
        # Move to next node (which is now in prev due to swap)
        curr = curr.prev
        position += 1
    
    # Swap head and tail
    print(f"\n  Swapping head and tail pointers...")
    dll.head, dll.tail = dll.tail, dll.head
    print(f"  New head: {dll.head.data}, New tail: {dll.tail.data}")
    
    print(f"\nList after reversal (pointer swap):")
    dll.print_forward("  ")
    dll.print_backward("  ")
    
    return dll


# ===== TEST CASES =====

if __name__ == "__main__":
    # ===== TEST CASE 1: Stack Approach =====
    print("\n" + "" + "=" * 68 + "")
    print(" " + " " * 66 + " ")
    print(" " + "TEST CASE 1: REVERSING USING STACK (Your Naive Approach)".center(66) + " ")
    print(" " + " " * 66 + " ")
    print("" + "=" * 68 + "")
    
    dll1 = DoublyLinkedList()
    elements = [10, 20, 30, 40, 50]
    
    print("\nOriginal List:")
    for elem in elements:
        dll1.append(elem)
    
    dll1.print_forward("  ")
    dll1.print_backward("  ")
    
    # Reverse using stack
    dll1_reversed = reverse_dll_stack(dll1)
    
    print("\nAfter Reversal:")
    dll1_reversed.print_forward("  ")
    dll1_reversed.print_backward("  ")
    
    
    # ===== TEST CASE 2: Pointer Swap Approach (Optimal) =====
    print("\n\n" + "" + "=" * 68 + "")
    print(" " + " " * 66 + " ")
    print(" " + "TEST CASE 2: REVERSING USING POINTER SWAP (OPTIMAL)".center(66) + " ")
    print(" " + " " * 66 + " ")
    print("" + "=" * 68 + "")
    
    dll2 = DoublyLinkedList()
    elements = [10, 20, 30, 40, 50]
    
    print("\nOriginal List:")
    for elem in elements:
        dll2.append(elem)
    
    dll2.print_forward("  ")
    dll2.print_backward("  ")
    
    # Reverse using pointer swap
    dll2_reversed = reverse_dll_pointers(dll2)
    
    print("\nAfter Reversal:")
    dll2_reversed.print_forward("  ")
    dll2_reversed.print_backward("  ")
    
    
    # ===== TEST CASE 3: Different List Size =====
    print("\n\n" + "" + "=" * 68 + "")
    print(" " + " " * 66 + " ")
    print(" " + "TEST CASE 3: REVERSING LIST WITH 7 ELEMENTS".center(66) + " ")
    print(" " + " " * 66 + " ")
    print("" + "=" * 68 + "")
    
    dll3 = DoublyLinkedList()
    elements = [1, 2, 3, 4, 5, 6, 7]
    
    print("\nOriginal List:")
    for elem in elements:
        dll3.append(elem)
    
    dll3.print_forward("  ")
    dll3.print_backward("  ")
    
    # Reverse using pointer swap
    dll3_reversed = reverse_dll_pointers(dll3)
    
    print("\nAfter Reversal:")
    dll3_reversed.print_forward("  ")
    dll3_reversed.print_backward("  ")
    
    
    # ===== COMPLEXITY COMPARISON =====
    print("\n\n" + "" + "=" * 68 + "")
    print(" " + " " * 66 + " ")
    print(" " + "COMPLEXITY COMPARISON".center(66) + " ")
    print(" " + " " * 66 + " ")
    print("" + "=" * 68 + "")
    
    print(f"\n{'Approach':<35} {'Time':<20} {'Space':<20}")
    print("" * 75)
    print(f"{'Stack (Naive)':<35} {'O(n)':<20} {'O(n)':<20}")
    print(f"{'Pointer Swap (Optimal)':<35} {'O(n)':<20} {'O(1)':<20}")
    print("" * 75)
    print("\n BEST APPROACH: Pointer Swap - O(n) time, O(1) space")
    print("  (No extra space needed, just swapping pointers)")
```

### `Approach`
- Initializing emypty arr 
- In first iterationn jumping temp pointer 2 steps starting from head pointer till None is not found and storing it into arr.
- Now jumping head head pointer from 2nd node till null to counter even node and storing it into arr.

```python
from typing import Optional

# Definition for singly linked list
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution:
    def oddEvenList(self,head):
        """
        Reverse-engineer approach using array:
        1. Store odd-indexed nodes in array
        2. Store even-indexed nodes in array
        3. Replace values sequentially
        
        Time Complexity: O(n)
        Space Complexity: O(n) - array storage
        """

        # Edge Case: Empty List or single node
        if not head or not head.next:
            return head
        
        arr = []

        # Step 1: Collect odd-indexed nodes (1st, 3rd, 5th, ...)
        temp = head
        while temp:
            arr.append(temp.val)
            if temp.next:
                temp = temp.next.next    # Jump 2 steps
            else:
                break
        
        # Step 2: Collect even-indexed nodes (2nd, 4th, 6th, ...)
        temp = head.next
        while temp:
            arr.append(temp.val)
            if temp.next:
                temp = temp.next.next   # Jump 2 steps
            else:
                break
        
        # Step 3: Replace original list values with array values
        curr = head
        idx = 0
        while curr and idx < len(arr):
            curr.val = arr[idx]
            curr = curr.next
            idx += 1
        
        return head
```

```python
# Helper Functions
def create_linked_list(arr):
    """Create linked list from array"""
    if not arr:
        return None
    head = ListNode(arr[0])
    curr = head
    for val in arr[1:]:
        curr.next = ListNode(val)
        curr = curr.next
    return head

def print_linked_list(head):
    """Print linked list"""
    values = []
    curr = head
    while curr:
        values.append(str(curr.val))
        curr = curr.next
    return " -> ".join(values)
```

```python
# ===== TEST CASES WITH EDGE CASES =====

if __name__ == "__main__":
    solution = Solution()
    
    print("=" * 70)
    print("ODD EVEN LINKED LIST - ARRAY-BASED APPROACH")
    print("=" * 70)

    # Edge Case 1: Empty list (None)
    print("\n\nEDGE CASE 1: Empty list (None)")
    print("─" * 70)
    head1 = None
    print(f"Original: None")
    head1 = solution.oddEvenList(head1)
    print(f"Result:   {print_linked_list(head1) if head1 else 'None'}")
    print(f"Expected: None")
    
    
    # Edge Case 2: Single node
    print("\n\nEDGE CASE 2: Single node [5]")
    print("─" * 70)
    head2 = create_linked_list([5])
    print(f"Original: {print_linked_list(head2)}")
    head2 = solution.oddEvenList(head2)
    print(f"Result:   {print_linked_list(head2)}")
    print(f"Expected: 5")

    # Edge Case 3: Two nodes
    print("\n\nEDGE CASE 3: Two nodes [1, 2]")
    print("─" * 70)
    head3 = create_linked_list([1, 2])
    print(f"Original: {print_linked_list(head3)}")
    head3 = solution.oddEvenList(head3)
    print(f"Result:   {print_linked_list(head3)}")
    print(f"Expected: 1 -> 2")
    
    # Test Case 1: Standard list with 5 nodes
    print("\n\nTEST CASE 1: Standard list [1, 2, 3, 4, 5]")
    print("─" * 70)
    head4 = create_linked_list([1, 2, 3, 4, 5])
    print(f"Original: {print_linked_list(head4)}")
    print(f"Step 1 - Odd indices (1,3,5):   [1, 3, 5]")
    print(f"Step 2 - Even indices (2,4):    [2, 4]")
    print(f"Step 3 - Array merged:          [1, 3, 5, 2, 4]")
    head4 = solution.oddEvenList(head4)
    print(f"Result:   {print_linked_list(head4)}")
    print(f"Expected: 1 -> 3 -> 5 -> 2 -> 4")

```

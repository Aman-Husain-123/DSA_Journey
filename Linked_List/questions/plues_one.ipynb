```python
class ListNode:
    def __init__(self,x):
        self.val = x
        self.next = None
```

```python
class solution:
    def detectCycle(self,head):
        visited = set()
        curr = head
        step = 0

        while curr:
            print(f"Step {step}: Visiting node with value {curr.val}")
            if curr in visited:
                print(f"Cycle detected at node with value {curr.val}")
                return curr
            visited.add(curr)
            curr = curr.next
            step += 1

        print("No cycle detected in the LL")
        return None

```

```python
# Example Usage
if __name__ == '__main__':
    # Create Node
    nodes = [ListNode(i) for i in range(1,6)]

    # Link nodes: 1 -> 2 -> 3 -> 4 -> 5
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i+1]
    
    # Create a cycle for testing: node 5 points back to node 2
    nodes[-1].next = nodes[1]

    solution = solution()
    cycle_start = solution.detectCycle(nodes[0])
    if cycle_start:
        print(f"Start of cycle is node with value: {cycle_start.val}")
    else:
        print("No cycle found.")
```

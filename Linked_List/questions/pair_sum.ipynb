```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
        
```

```python
def find_pairs(head,target):
    pairs = []
    first = head

     # Outer loop: traverse from head to end
    while first:
        second = first.next
        # Inner loop: check all nodes after first
        while second:
            if first.data + second.data == target:
                pairs.append( (first.data, second.data) )
            second = second.next
        first = first.next
    return pairs
```

```python
# Example usage:
# Creating DLL: 1 <-> 2 <-> 3 <-> 4 <-> 5
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)
n5 = Node(5)
n1.next = n2; n2.prev = n1
n2.next = n3; n3.prev = n2
n3.next = n4; n4.prev = n3
n4.next = n5; n5.prev = n4

print(find_pairs(n1, 5))  # Output: [(1, 4), (2, 3)]
```

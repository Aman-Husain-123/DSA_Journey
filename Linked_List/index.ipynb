```python
# TODO: Converting array into linkedlist
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

def convert_arr_to_ll(arr):
    head = Node(arr[0])
    mover = head
    for i in range(1, len(arr)):
        temp = Node(arr[i])
        mover.next = temp
        mover = temp
    return head

arr = [12, 5, 6, 8]
head = convert_arr_to_ll(arr)
print(head.data)
```

```python
# TODO: Length of the linked list
def lenght_of_ll(head):
    count = 0
    curr = head
    while curr:
        count += 1
        curr = curr.next
    return count
```

```python
# TODO: Searching an element in the linked list
def search_element(head,key):
    curr = head 
    while curr:
        if curr.data == key:
            return True
        curr = curr.next
    return False
```

### `Linked List Creation`

```python
class Node:
    def __init__(self,data = None,next = None):
        self.data = data
        self.next = next


class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_beginning(self,data):
        node = Node(data,self.head)
        self.head = node
    
    def print(self):
        if self.head is None:
            print("Linked List Is Empty")
            return
        
        itr = self.head
        llstr = ''

        while itr:
            llstr += str(itr.data) + ' --> '
            itr = itr.next
        
        print(llstr)
    
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data,None)
            return
    
        itr = self.head
        while itr.next:
            itr = itr.next
        
        itr.next = Node(data,None)
    
    def insert_values(self,data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
    
    def get_length(self):
        count = 0
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
        return count
    
    def remove_at(self,index):
        if index < 0 or index >= self.get_length():
            raise Exception("Invalid Index")
        
        if index == 0:
            self.head = self.head.next
            return
        
        count = 0
        itr = self.head
        while itr:
            if count == index - 1:
                itr.next = itr.next.next
                break
            itr = itr.next
            count += 1
```

### `Creating a node PW`

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
# Create a LL
class LinkedList:
    def __init__(self):
        self.head = None
    

    # Adding a node to linked list
    def add_node(self,value):
        new_node = Node(value)
        

        # Creating first(head) node
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
```

```python
# Display a LL
l1 = LinkedList()

l1.add_node(10)
l1.add_node(50)
l1.add_node(20)
```

### `Display a LL`

```python
# Iterative
def print_linked_list(head):
    curr = head
    while curr is not None:
        print(curr.data," -> ",end = "")
        curr = curr.next
```

```python
print_linked_list(l1.head)
```

```python
# Recursive
def print_recursiverly(head):
    if head is None:
        return

    print(head.data)
    print_recursiverly(head.next)

print_recursiverly(l1.head)
```

### `Finding the length of the LL`

```python
l2 = LinkedList()

for i in range(10):
    l2.add_node(i*2)
```

```python
print_linked_list(l2.head)
```

```python
# TODO: Finding length iteratively
def find_length(head):
    length = 0
    while head is not None:
        length += 1
        head = head.next
    
    return length
```

```python
find_length(l2.head)
```

```python
# TODO: Finding length recursively
def find_recursively(head):
    if head is None:
        return 0
    
    return 1 + find_recursively(head.next)
```

```python
find_recursively(l2.head)
```

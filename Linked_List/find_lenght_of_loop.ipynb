```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def loopLength(head):
    nodes = []
    curr = head
    
    while curr:
        if curr in nodes:  # Loop detected, curr repeats
            first_index = nodes.index(curr)
            loop_len = len(nodes) - first_index
            return loop_len
        nodes.append(curr)
        curr = curr.next
    
    return 0  # No loop

# Example usage
if __name__ == "__main__":
    # Creating linked list with loop: 1->2->3->4->5->6->2 ...
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)
    head.next.next.next.next = Node(5)
    head.next.next.next.next.next = Node(6)
    head.next.next.next.next.next.next = head.next  # Creates loop back to node with data=2
    
    length = loopLength(head)
    print(f"Length of loop is: {length}")
```

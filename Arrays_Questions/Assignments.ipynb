```python
# Given an array, predict if the array contains duplicates or not.
arr = [2, 3, 1]
dictt = {}

found = False

for num in arr:
    if num in dictt:
        found = True
        break
    dictt[num] = 1

if found:
    print("Duplicate")
else:
    print("No Duplicate")
```

```python
# TODO: Optimal
arr = [1, 2, 3, 5, 6]
lo = 0
hi = len(arr) - 1
smalles_element = float('-inf')

while lo <= hi:
    mid = (lo + hi) // 2

    if arr[mid] == mid + 1:
        lo = mid + 1
    else:
        hi = mid - 1

smalles_element = lo + 1
smalles_element
```

```python
# Write a program to reverse the array without using any extra array
arr = [1,2,3,4,5]
i,j = 0,len(arr)-1

while i <= j:
    temp = arr[i]
    arr[i] = arr[j]
    arr[j] = temp

    i += 1
    j -= 1
```

```python
# Rotate the given array â€˜aâ€™ by k steps, where k is non-negative.
def rotate_left(arr,k):
    # Handle empty array
    if not arr:
        return arr

    n = len(arr)
    # k = k % n

    def reverse(l,r):
        while l < r:
            arr[l],arr[r] = arr[r],arr[l]
            l += 1
            r -= 1
     
    # step 1: reverse first k elements
    reverse(0,n-k-1)

    # step 2: reverse remaining elements
    reverse(n-k,n-1)

    # step 3: reverse entire array
    reverse(0, n - 1)

    return arr           
```

```python
arr = [1, 2, 3, 4, 5, 6, 7]
k = 3

rotate_left(arr, k)
print(arr)
```

```python
# Segregate 0s and 1s
arr = [0,0,1,1,0]
noz = 0

for i in range(len(arr)):
    if arr[i] == 0:
        noz += 1

for i in range(len(arr)):
    if i < noz:
        arr[i] = 0
    else:
        arr[i] = 1
        
```

```python
# Move all negative numbers to beginning and positive to end with constant extra space
arr = [1,8,3,6,-5,-4,-3]
n = len(arr)

left,right = 0,n-1
while left < right:
    if arr[left] < 0:
        left += 1
    elif arr[right] > 0:
        right -= 1
    else:
        arr[left],arr[right] = arr[right],arr[left]
arr
```

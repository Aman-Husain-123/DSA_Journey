```python
arr = [2, 4, 5, 7, 3]
prefix_sum = [0]

for num in arr:
    prefix_sum.append(prefix_sum[-1] + num)
print(prefix_sum)
```

```python
arr = [2, 4, 5, 7, 3]
n = len(arr)
prefix_sum = [0] * n

prefix_sum[0] = arr[0]

for i in range(1,len(arr)):
    prefix_sum[i] = prefix_sum[i-1] + arr[i]
print(prefix_sum)
```

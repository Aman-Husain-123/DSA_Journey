```python
def longest_successive_elements(a: list) -> int:
   
    n = len(a)
    
    # Check for empty array (equivalent to 'if (n==0) return 0;')
    if n == 0:
        return 0

    # 1. Store all elements in a set for O(1) average time complexity lookups.
    num_set = set(a)
    longest = 1

    # 2. Iterate over the elements in the set (equivalent to 'for(auto it : st)').
    # This ensures each sequence is processed only once from its starting element.
    for x in num_set:
        
        # Check if 'x' is the starting element of a sequence:
        # If (x - 1) is NOT found in the set, then 'x' is the start.
        # (Equivalent to 'if(st.find(it - 1) == st.end())')
        if (x - 1) not in num_set:
            
            # Start counting this new sequence
            current_num = x
            current_count = 1  # Equivalent to 'int cnt = 1;'
            
            # 3. Check for successive elements
            # (Equivalent to 'while(st.find(x+1) != st.end())')
            while (current_num + 1) in num_set:
                current_num += 1
                current_count += 1
            
            # 4. Update the maximum length found
            # (Equivalent to 'longest = max(longest, cnt);')
            longest = max(longest, current_count)

    return longest

# Example Usage
arr = [100, 4, 200, 1, 3, 2]
result = longest_successive_elements(arr)

print(f"Input Array: {arr}")
print(f"Longest Consecutive Sequence Length: {result}")
```

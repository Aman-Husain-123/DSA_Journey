```python
# Brute Force (Element Should Appear More Than (n / 2) times)

def func(arr):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(n):
            if arr[i] == arr[j]:
                count += 1
        
        if (count > (n / 2)):
            return arr[i]

```

```python
arr = [1,2,2,2,2,1,3,3,3,3,3,3,6,4]
func(arr)
```

```python
# Better Solution (Hashing)  N(logN) + O(N)
def majorityelement(v):
    mpp = {}

    for num in v:
        mpp[num] = mpp.get(num,0) + 1
    
    print(mpp)
    for key,value in mpp.items():
        if value > len(v) // 2:
            return key
    
    return -1
```

```python
# Example usage:
v = [2, 2, 1, 1, 1,1,1, 2, 2]
print(majorityelement(v))  # Output: 2
```

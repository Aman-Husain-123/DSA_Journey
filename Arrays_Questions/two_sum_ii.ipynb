```python
# Brute Force Approach to find two indices such that their values sum up to a target
arr = [2,71,11,15]
target = 9
for i in range(len(arr)):
    for j in range(i + 1,len(arr)):
        if arr[i] + arr[j] == target:
            print([i,j])
```

```python
# Better Approach using Hash Map
arr = [2,7,11,15]
target = 9
hash_map = {}
for i in range(len(arr)):
    complement = target - arr[i]
    if complement in hash_map:
        print([hash_map[complement], i])    ## Key :- complement and Value :- hash_map[complement] for example 0:2
    hash_map[arr[i]] = i      ## Key :- arr[i] and Value :- i for example 2:0, 7:1

print(hash_map)
```

```python
# Optmized Approach using Two Pointers
arr = [2,7,11,15]
target = 9
left = 0
right = len(arr) - 1

for i in range(len(arr)):
    curr_sum = arr[left] + arr[right]
    if curr_sum == target:
        print([left, right])
        break
    elif curr_sum < target:
        left += 1
    else:
        right -= 1
    
print(-1)
```

```python
# TC :- O(2N)
# SC :- O(N)

arr = [1, 4, 0, 3, 2, 0]
temp = []
n = len(arr)
print("Original Array:",arr)

# Step 1: Collect non-zero elements
for i in range(n):
    if arr[i] != 0:
        temp.append(arr[i])

print(f"Temporary Array:{temp}")

# Step 2: Copy back non-zero elements
for i in range(len(temp)):
    arr[i] = temp[i]

print(f"Copying Back Non-Zero Element:{arr}")

# Step 3: Fill the rest with zeros
for i in range(len(temp), n):
    arr[i] = 0


print(f"Final Array:{arr}")  # âœ… [1, 4, 3, 2, 0, 0]
```

```python
# TODO: Brute Force : O(n^2)
arr = [17,18,5,4,6,1]
ans = []
n = len(arr)
for i in range(n):
    leader = True
    for j in range(i + 1,n):
        if arr[j] > arr[i]:
            leader = False
            break
    
    if leader == True:
        ans.append(arr[i])

ans
```

```python
# TODO: Optimal Approach
arr = [17,18,5,4,6,1]
maxx = float('-inf')
n = len(arr)
ans = []
for i in range(n-1,-1,-1):
    if arr[i] > maxx:
        ans.append(arr[i])
    
    maxx = max(maxx,arr[i])

ans = ans[::-1]
print(ans)
```

```python
arr = [17,18,5,4,6,1]
n = len(arr)
max_so_far = -1   # last element must be -1
# Traverse from right to left
for i in range(n-1, -1, -1):
    temp = arr[i]          # store current element
    arr[i] = max_so_far    # replace with greatest to right
    max_so_far = max(max_so_far, temp)  # update max
print(arr)
```

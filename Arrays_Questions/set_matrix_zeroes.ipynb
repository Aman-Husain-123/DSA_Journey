```python
matrix = [[0,1,2,0],[3,4,5,2],[1,3,1,5]]

rows = len(matrix)
cols = len(matrix[0])

def markrow(matrix, row, cols):
    for j in range(cols):
        if matrix[row][j] != 0:
            matrix[row][j] = -1

def markcol(matrix, col, rows):
    for i in range(rows):
        if matrix[i][col] != 0:
            matrix[i][col] = -1

# First pass: mark rows and columns
for i in range(rows):
    for j in range(cols):
        if matrix[i][j] == 0:
            markrow(matrix, i, cols)
            markcol(matrix, j, rows)

# Second pass: set all -1 to 0
for i in range(rows):
    for j in range(cols):
        if matrix[i][j] == -1:
            matrix[i][j] = 0

# Print the result
for row in matrix:
    print(row)
```

```python
matrix = [[1,1,1],[1,0,1],[1,1,1]]

firstrow, firstcol = False, False

# Set markers in first row and column
for i in range(len(matrix)):
    for j in range(len(matrix[0])):
        if matrix[i][j] == 0:
            if i == 0:
                firstrow = True
            if j == 0:
                firstcol = True
            matrix[0][j] = 0
            matrix[i][0] = 0

# Replace inner matrix (skip first row and column)
for i in range(1, len(matrix)):
    for j in range(1, len(matrix[0])):
        if matrix[i][0] == 0 or matrix[0][j] == 0:
            matrix[i][j] = 0

# Last remaining checks
if firstrow:
    for j in range(len(matrix[0])):
        matrix[0][j] = 0

if firstcol:
    for i in range(len(matrix)):
        matrix[i][0] = 0

# Print the result
for row in matrix:
    print(row)
```

```python
# TC :- O (n + d)
# SC :- O (d)

from typing import List
def left_rotate(arr,n,d) -> List:
    d = d % n    # Reduce d lesser than n
    print(d)
    temp = []

    # TODO: Storing the rotated part of elements in temp array
    for i in range(0,d):
        temp.append(arr[i])
    
    # TODO: Then Shifting
    for i in range(d,n):
        arr[i - d] = arr[i]
    
    # TODO: Replacing Back
    for i in range(n - d,n):
        arr[i] = temp[i - (n - d)]
    
    return arr
```

```python
arr = [1,2,3,4,5,6,7]
n = len(arr)
d = 3
print(left_rotate(arr,n,d))
```

### Optimal Approach

```python
from typing import List

def reverse(arr: List[int], start: int, end: int) -> None:
    """Reverse elements of arr from index start to end (inclusive) in-place."""
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1

def left_rotate(arr: List[int], n: int, d: int) -> List[int]:
    d = d % n  # handle case when d > n
    
    # Step 1: Reverse first d elements
    reverse(arr, 0, d - 1)
    
    # Step 2: Reverse the rest
    reverse(arr, d, n - 1)
    
    # Step 3: Reverse the entire array
    reverse(arr, 0, n - 1)
    
    return arr


# Example
arr = [1, 2, 3, 4, 5, 6, 7]
n = len(arr)
d = 3
print(left_rotate(arr, n, d))  # âœ… [4, 5, 6, 7, 1, 2, 3]
```

```python
def twoSum(nums, target):

    mpp = {} 
    
    for i in range(len(nums)):
        num = nums[i]
        moreNeeded = target - num
        if moreNeeded in mpp:
            return [mpp[moreNeeded], i]
        mpp[num] = i
        print(mpp)
    return []

# Example Usage:
nums = [2, 7, 11, 15]
target = 9
result = twoSum(nums, target)
print(f"Indices for {nums} and target {target}: {result}") # Output: Indices for [2, 7, 11, 15] and target 9: [0, 1]

```

```python
arr = [3, 1, -2, -5, 2, -4]

pos_arr = []
neg_arr = []

# Separate positives and negatives
for num in arr:
    if num >= 0:
        pos_arr.append(num)
    else:
        neg_arr.append(num)

# Merge them alternately
i = j = k = 0
n = len(arr)

while i < len(pos_arr) and j < len(neg_arr):
    if k % 2 == 0:  # even index → positive
        arr[k] = pos_arr[i]
        i += 1
    else:           # odd index → negative
        arr[k] = neg_arr[j]
        j += 1
    k += 1

# Fill leftover positives
while i < len(pos_arr):
    arr[k] = pos_arr[i]
    i += 1
    k += 1

# Fill leftover negatives
while j < len(neg_arr):
    arr[k] = neg_arr[j]
    j += 1
    k += 1

print(arr)
```

```python
arr = [3, 1, -2, -5, 2, -4]
n = len(arr)

# Step 1: Partition into positives and negatives
i, j = 0, n - 1
while i < j:
    while i < n and arr[i] > 0:  # move until you find a negative
        i += 1
    while j >= 0 and arr[j] < 0:  # move until you find a positive
        j -= 1
    if i < j:
        arr[i], arr[j] = arr[j], arr[i]

# Step 2: Rearrange in alternate positions
pos, neg = 0, i   # positives from 0..i-1, negatives from i..n-1
while pos < n and neg < n and arr[pos] > 0:
    arr[pos + 1], arr[neg] = arr[neg], arr[pos + 1]
    pos += 2
    neg += 1

print(arr)
```

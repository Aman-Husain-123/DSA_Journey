```python
arr = [1,2,3,4,1,2,6,7,9]
lenn = 0  # safe name, not overriding len()

for i in range(len(arr)):    
    for j in range(i, len(arr)):
        s = 0
        for k in range(i, j+1):   # include j
            s += arr[k]
        if s == 0:                # check sum
            lenn = max(lenn, j - i + 1)

print(lenn)
```

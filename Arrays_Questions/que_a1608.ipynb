```python
def square(arr):
    n = len(arr)
    for x in range(1,n + 1):
        count = sum(num >= x for num in arr)
        if count == x:
            return x
            
    return -1 
```

```python
arr = [0,0,1,2,3,9,9]
print(square(arr))
```

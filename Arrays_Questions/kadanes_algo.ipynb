```python
from typing import List
def maxSubArray(nums: List[int]) -> int:
        maximum = float('-inf')
        n = len(nums)
        for i in range(n):
            for j in range(i,n):
                summ = 0
                for k in range(i,j):
                    summ += nums[k]
                
                maximum = max(summ,maximum)

        return maximum
```

```python
nums = [-2,-3,4,-1,-2,1,5,-3]
maxSubArray(nums)
```

```python
nums = [-2, -3, 4, -1, -2, 1, 5, -3]

max_sum = float('-inf')
curr_sum = 0
start = 0
ans_start, ans_end = 0, 0

for i in range(len(nums)):
    curr_sum += nums[i]

    if curr_sum > max_sum:
        max_sum = curr_sum
        ans_start = start
        ans_end = i

    if curr_sum < 0:
        curr_sum = 0
        start = i + 1  # reset start index

print("Maximum Subarray Sum:", max_sum)
print("Subarray indices:", ans_start, ans_end)
print("Subarray:", nums[ans_start:ans_end+1])
```

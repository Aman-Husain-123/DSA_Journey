```python
def ternary_search(arr,i,j,key):
    mid1 = i + (j - i) // 3
    mid2 = j - (j - i) // 3

    while i <= j:
        if arr[mid1] == key:
            return mid1
        
        if arr[mid2] == key:
            return mid2
        
        elif arr[mid1] > key:
            return ternary_search(arr,i,mid1-1,key)
        
        elif arr[mid1] < key and arr[mid2] > key:
            return ternary_search(arr,mid1 + 1,mid2-1,key)
        
        elif arr[mid2] < key:
            return ternary_search(arr,mid2 + 1,j,key)
    
    return -1
```

```python
arr = [12,23,33,35,39,45,67,78,90]
i,j = 0,len(arr) - 1
key = 90
ternary_search(arr,i,j,key)
```

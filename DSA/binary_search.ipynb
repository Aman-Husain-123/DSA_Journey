```python
def binary_search_recursive(arr, target, low, high):
    # Base case: if range is invalid, target not found
    if low > high:
        return -1

    mid = low + (high - low) // 2   # If value of high is too large it would be diff for variable to store

    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        # Search in left half
        return binary_search_recursive(arr, target, low, mid - 1)
    else:
        # Search in right half
        return binary_search_recursive(arr, target, mid + 1, high)


# Example usage
arr = [5, 21, 45, 212, 4343]  # must be sorted
target = 45
result = binary_search_recursive(arr, target, 0, len(arr) - 1)

print("Element found at index:", result if result != -1 else "Not Found")
```

### Using Recursion

```python
def binary_search_recursive_1(arr,target,i,j):
        if i > j:   # Base Case
            return -1
        
        mid = i + (j - i) // 2

        # Mid element is the target element
        if arr[mid] == target:
            return mid
        
        elif arr[mid] > target:
            # Search space - i to mid - 1 (Left side of the mid)
            return binary_search_recursive_1(arr,target,i,mid-1)
        
        else:
            # Search space - mid + 1 to j (Right side of the mid)
            return binary_search_recursive_1(arr,target,mid + 1,j)
    
```

```python
# Driver code
arr = [12,22,32,43,54,55]
target = 54
i = 0
j = len(arr) - 1
print(binary_search_recursive_1(arr,target,i,j))
```

### Without Using Recursion

```python
def binary_search_recursive_2(arr,target,i,j):
    while i <= j:
        mid = i + (j - i) // 2

        # Mid element is the target element
        if arr[mid] == target:
            return mid
        
        elif arr[mid] > target:
            # Search space - i to mid - 1 (Left side of the mid)
            i = mid - 1
        
        else:
            # Search space - mid + 1 to j (Right side of the mid)
            i = mid + i
    return -1
    
```

```python
# Driver code
arr = [12,22,32,43,54,55]
target = 32
i = 0
j = len(arr) - 1
print(binary_search_recursive_2(arr,target,i,j))
```

```python
#### Que :- 1
def first_infinite_recursive(arr, low, high):
    # Base case: no ∞ found
    if low > high:
        return -1

    mid = (low + high) // 2

    if arr[mid] == float("inf"):
        # Check if it's the first ∞
        if mid == 0 or arr[mid - 1] != float("inf"):
            return mid
        else:
            # Look on the left side
            return first_infinite_recursive(arr, low, mid - 1)
    else:
        # Look on the right side
        return first_infinite_recursive(arr, mid + 1, high)


# Driver code
arr = [7, 12, 3, 15, 21, float("inf"), float("inf"), float("inf")]
print(first_infinite_recursive(arr, 0, len(arr) - 1))  # Output: 5
```

```python
### Que :- 2
def search(arr,target,i,j):
    while i <= j:
        if arr[i] + arr[j] == target:
            return (i,j)
        elif arr[i] + arr[j] > target:
            j -= 1
        else:
            i += 1
```

```python
arr = [12,23,45,67,89,99]
target = 188
i = 0
j = len(arr) - 1
print(search(arr,target,i,j))
```

```python
# Que :3 Buy and sell stock O(n)

def stock_price(price):
    maxProfit = 0
    buyPrice = float('inf')
    sellingPrice = 0

    for i in range(len(price) - 1):
        if buyPrice < price[i]:       # Profit
            profit = price[i] - buyPrice    # Todays Profit
            if profit > maxProfit:
                maxProfit = profit
        else:
            buyPrice = price[i]
    return maxProfit
```

```python
price = [7,1,5,3,9,4]
print(stock_price(price))
```

```python
# Que :- 4
def searched(arr,target,i,j,n):
    while i <= j:
        mid = i + (j - i) // 2
        if arr[mid] == target:
            row_number = mid // n
            col_number = mid % n
            return (row_number, col_number)
        elif arr[mid] > target:
            i = mid - 1
        else:
            i = mid  + 1

    return -1
```

```python
arr = [12,34,45,65,76,77,89,99,101]  # Flattened 3x3 matrix
print(searched(arr, 65, 0, len(arr)-1, 3))
```

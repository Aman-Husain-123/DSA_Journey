### Interview Questions

### 1. Reverse a List
Problem: Write a Python function that reverses a list without using the reversed() function.

```python
L = [1, 2, 3, 4, 5]
length = len(L)

for i in range(length - 1, -1, -1):  # Loop from last index to 0
    print(L[i], end=",")
```

### 4. Check for Duplicates
Problem: Write a Python function that checks whether a list contains any duplicates.

```python
L = [1, 2, 3, 4, 5,1,1]
count = 0
for i in L:
    if len(set(L)) == len(L):
        count += 1

if count == 1:
    print('No')
else:
    print('Yes')
```

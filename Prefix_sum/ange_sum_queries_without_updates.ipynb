### `Before Prefix Sum`

```python
# Function to compute sum in given range
def rangeSum(arr,i,j):
    sum = 0

    # Compute sum from i - j
    for k in range(1,j + 1):
        sum += arr[k]
    
    return sum
```

```python
# Driver code
if __name__ == "__main__" :
  arr = [ 1, 2, 3, 4, 5 ]
  print(f"Array : {arr}")

  print("Index : [0, 1, 2, 3, 4] \n")
  print(f"Sum between index 1-3 is : {rangeSum(arr, 1, 3)}")

  print(f"Sum between index 2-4 is : {rangeSum(arr, 2, 4)}")
```

### `After Prefix Sum`

```python
# [Expected Solution] - Prefix Sum - O(1) for Every Query and O(n) Space

# Function to precompute the prefix sum

def pre_compute(arr):
    pre = [0] * len(arr)
    pre[0] = arr[0]
    for i in range(1,len(arr)):
        pre[i] = arr[i] + pre[i - 1]
    
    return pre
```

```python
# Returns sum of elements in arr[i..j]
# It is assumed that i <= j

def rangeSum(i,j,pre):
    if i == 0:
        return pre[j]
    return pre[j] - pre[i - 1]
```

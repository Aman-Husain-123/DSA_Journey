```python
arr = [1, 2, 3, 7]
target = 12
n = len(arr)
start = 0
curr_sum = 0
ans = []

for end in range(n):
    curr_sum += arr[end]

    # shrink from left while sum is too big
    while curr_sum > target and start <= end:
        curr_sum -= arr[start]
        start += 1

    if curr_sum == target:
        ans.append([start+1,end+1])
```

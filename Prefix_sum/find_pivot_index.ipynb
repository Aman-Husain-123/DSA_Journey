```python
nums = [1, 7, 3, 6, 5, 6]

total_sum = 0
for i in range(len(nums)):
    total_sum += nums[i]

left_sum = 0
ans = -1

for i in range(len(nums)):
    right_sum = total_sum - nums[i] - left_sum

    if left_sum == right_sum:
        ans = i
        break

    left_sum += nums[i]

if ans == -1:
    print("Not Found")
else:
    print(f"Found at {ans}")
```

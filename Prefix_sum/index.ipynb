```python
# Prefix Sum
arr = [10,20,30,40]
prefix_sum = [0] * len(arr)

for i in range(1,len(arr)):
    prefix_sum[i] = prefix_sum[i-1] + arr[i-1]
    # print(f"{prefix_sum[i-1]} + {arr[i-1]} = " , prefix_sum[i])
```

```python
# Suffix Sum
arr = [3,9,2,5,6]
suffix_sum = [0] * len(arr)
n = len(arr)

for i in range(n-2,-1,-1):
    suffix_sum[i] = suffix_sum[i+1] + arr[i+1]
suffix_sum
```

```python
# Input: arr[] = [10, 3, 5, 6, 2]
# Output: [180, 600, 360, 300, 900]
# Explanation: For i=0, res[i] = 3 * 5 * 6 * 2 is 180.
# For i = 1, res[i] = 10 * 5 * 6 * 2 is 600.
# For i = 2, res[i] = 10 * 3 * 6 * 2 is 360.
# For i = 3, res[i] = 10 * 3 * 5 * 2 is 300.
# For i = 4, res[i] = 10 * 3 * 5 * 6 is 900
```

```python
nums = [10, 3, 5, 6, 2]
n = len(nums)
pre = [0] * n
suff = [0] * n
ans = [0] * n

# # Prefix prod array
p = nums[0]
pre[0] = 1
for i in range(1,n):
    pre[i] = p
    p *= nums[i]
        
# # Suffix produxt array
p = nums[n-1]
suff[n-1] = 1
for i in range(n-2,-1,-1):
    suff[i] = p
    p *= nums[i]

# making ans in prefix array
for i in range(n):
    pre[i] = pre[i] * suff[i]
        
print(pre)
```

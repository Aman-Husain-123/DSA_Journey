```python
arr = [10, 20, 30, 40, 50]
count = 1
n = len(arr)
prefix_sum = [0] * len(arr)
prefix_sum[0] = arr[0]
        
for i in range(1,n):
    prefix_sum[i] = prefix_sum[i-1] + arr[i]
        
for i in range(len(prefix_sum)):
    prefix_sum[i] = prefix_sum[i] // count
    count += 1
    
prefix_sum
```

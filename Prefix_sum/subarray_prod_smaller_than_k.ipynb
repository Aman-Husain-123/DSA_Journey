```python
def numSubarrayProductLessThanK(nums,k) -> int:
        if k <= 1:
            return 0
        
        n = len(nums)
        i = 0
        j = 0
        count,product = 0,1
        while j < n:
            product *= nums[j]     # Prefix Prod
            while product >= k:  
                count += (j-i)
                product /= nums[i]
                i += 1
            j += 1
        
        while i < n:
            count += (j-i)
            product /= nums[i]
            i += 1
        return count
```

```python
arr = [10,5,2,6]
k = 100
print(numSubarrayProductLessThanK(arr,k))  # Output: 8
```

```python
arr = [10,5,2,6]
k = 100
count = 0

i,j = 0,0
n = len(arr)

for i in range(n):
    product = 1
    while j < n:
        product *= arr[j]
        if product < k:
            count += (j - i + 1)
            j += 1
        else:
            break
print(count)  # Output: 8
```

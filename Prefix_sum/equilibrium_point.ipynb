```python
arr = [1,2,6,4,0,-1]
left_sum = 0
point = -1
total = sum(arr)

for i in range(len(arr)):
    right_sum = total - left_sum - arr[i]

    if left_sum == right_sum:
        point = i
        break
    
    left_sum += arr[i]
print(point)
```

```python
arr = [3,4,8,2,1,4]
n = len(arr)
prefix_sum = [0] * n
suffix_sum = [0] * n

prefix_sum[0] = arr[0]
suffix_sum[n-1] = arr[n-1]

# Prefix Sum
for i in range(1,n):
    prefix_sum[i] = prefix_sum[i-1] + arr[i]


# Suffix Sum
for i in range(n-2,-1,-1):
    suffix_sum[i] = arr[i] + suffix_sum[i + 1]


diff = [0] * n
for i in range(n):
    diff[i] = prefix_sum[i] - suffix_sum[i]

count = -1
for i in range(len(diff)):
    if diff[i] == 0:
        count = i

if count == -1:
    print("No Equilibrium Point")
else:
    print(f"Equilibrium Point Is {count}")

print(prefix_sum)
print(suffix_sum)
print(diff)
```

```python
arr = [3,4,8,2,1,4,1]
n = len(arr)
prefix_sum = [0] * n
suffix_sum = [0] * n

prefix_sum[0] = arr[0]
suffix_sum[n-1] = arr[n-1]

# Prefix Sum
for i in range(1,n):
    prefix_sum[i] = prefix_sum[i-1] + arr[i]


# Suffix Sum
for i in range(n-2,-1,-1):
    suffix_sum[i] = arr[i] + suffix_sum[i + 1]


count = -1
for i in range(n):
    if prefix_sum[i] == suffix_sum[i]:
        count = i

if count == -1:
    print("No Equilibrium Point")
else:
    print(f"Equilibrium Point Is {count}")
```

```python
# TODO: Brute Approach
"""
TC :- O(n^2)
"""
arr = [1,1,1]
k = 2

count = 0

for start in range(len(arr)):
    curr_sum = 0
    for end in range(start, len(arr)):
        curr_sum += arr[end]
        
        # TODO: Print the subarray
        print(arr[start : end + 1])

        # TODO: Keeping the count of subarrays with sum equal to k
        if curr_sum == k:
            count += 1

print(count)
```

```python
# TODO: Optimized Approach
count = 0
nums = [1,1,1]
k = 2
prefix_sum = {}
right = 0 
curr_sum = 0

for _ in range(right,len(nums)):
            prefix_sum[curr_sum] = 1 + prefix_sum.get(curr_sum,0)
            curr_sum = curr_sum + nums[right]

            if (curr_sum - k) in prefix_sum:
                count = count + prefix_sum[curr_sum - k]
            
            right += 1
print(count)
```
